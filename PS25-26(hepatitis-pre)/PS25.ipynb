```python
import matplotlib.pyplot as plt
import seaborn as sns

def visBoxplots(df):
    numeric_cols = df.select_dtypes(include=['number']).columns
    cols = len(numeric_cols)
    
    rows = (cols + 3) // 4
    cols = min(cols, 4)
    
    fig,axes = plt.subplots(rows, cols, figsize=(16,4*rows))
    
    if rows > 1:
        axes = axes.flatten()
        
    for i,col in enumerate(numeric_cols):
        sns.boxplot(x=df[col],ax=axes[i])
        axes[i].set_title(col)
        
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
    
visBoxplots(df)
```

```python
print(df.shape)
def rem1(df):
    numeric_cols = df.select_dtypes(include=['number'])
    zscore = np.abs((numeric_cols - numeric_cols.mean()) / numeric_cols.std())
    out = zscore > 3
    df = df[~out.any(axis=1)]
    return df

def rem2f(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_threshold = Q1 - 1.5 * IQR
    upper_threshold = Q3 + 1.5 * IQR
    outliers = (df < lower_threshold) | (df > upper_threshold)
    print(outliers.sum())
    
def rem2(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_threshold = Q1 - 1.5 * IQR
    upper_threshold = Q3 + 1.5 * IQR
    outliers = (df < lower_threshold) | (df > upper_threshold)
    df = df[~((df < lower_threshold) | (df > upper_threshold)).any(axis=1)]
    return df
    
ndf = rem1(df)
df = ndf
print(df.shape)
visBoxplots(df)
```

# Model Building

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

df['Class'].unique()
```

```python
le = LabelEncoder()
df['Class'] = le.fit_transform(df['Class'])
```

```python
df.head()
```

```python
df['Class'].unique()
```

```python
X = df.drop('Class', axis =1)
Y = df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
modelLR = LogisticRegression()
modelLR.fit(X_train, Y_train)
```

```python
y_pred = modelLR.predict(X_test)
accuracy_score(Y_test,y_pred)
```

```python
modelNB = GaussianNB()
modelNB.fit(X_train,Y_train)
```

```python
y_pred = modelNB.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred)
```
